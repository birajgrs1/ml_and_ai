```python
#  importing libraries
import csv
import requests
from bs4 import BeautifulSoup
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}
```

```python
web_url = "https://hukut.com/mobile-phones"

response = requests.get(web_url, headers= headers)
print(response)
```

```python
# response.content
```

```python
soup = BeautifulSoup(response.content,"html.parser")
print(soup.prettify())
```

```python
# print(soup.body)
```

```python
data = []
```

```python
# Extract mobile names and prices for different brands
for ul in soup.find_all('ul'):
    for li in ul.find_all('li'):
        text = li.get_text(strip=True)
        if ' - ' in text:
            try:
                name, price = text.rsplit(' - ', 1)
                data.append([name, price])
            except ValueError:
                print(f"Warning: Could not split the text correctly: {text}")

csv_file_path = 'csv/mobile_prices.csv'
```

```python
with open(csv_file_path, 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Mobile Name', 'Price'])  # Write header
    writer.writerows(data)  # Write data

print(f"Data has been written to {csv_file_path}")
```
