```python
#  plotting heatmap 

plt.figure(figsize = (25,25))
sns.heatmap(df.isnull())
```

```python
null_var = df.isnull().sum()/df.shape[0] * 100   # converting in %
null_var
```

```python
drop_columns = null_var[null_var > 2].index

print(drop_columns)
```

```python
new_df = df.drop(columns = drop_columns)
```

```python
new_df.shape
```

```python
sns.heatmap(new_df.isnull())
```

```python
new_clear_df = new_df.dropna()
print(new_clear_df.shape)
```

```python
plt.figure(figsize = (25,25))
sns.heatmap(new_clear_df.isnull())
```

```python
#  Overlapping
sns.distplot(df['MSSubClass'])
sns.distplot(new_clear_df['MSSubClass'])
```

```python
num_var = ['Id', 'MSSubClass', 'LotArea', 'OverallQual', 'OverallCond',
           'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2',
           'BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF',
           'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath',
           'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces',
           'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF',
           'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal',
           'MoSold', 'YrSold', 'SalePrice']

plt.figure(figsize=(25, 25))
for i, var in enumerate(num_var):
    plt.subplot(9, 4, i + 1)
    sns.histplot(df[var], bins=20, kde=True)  
    sns.histplot(new_clear_df[var], bins=20, kde=True)
plt.tight_layout()
plt.show()
```
