```python
#  Importing necessary libraries 

import pandas as pd 
import numpy as np 
from sklearn.impute import SimpleImputer
```

```python
#  loading datasets

train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

print("Shape of train dataframe: ", train_df.shape)
print("Shape of test dataframe: ", test_df.shape)
```

```python
train_df.head(10)
```

```python
train_df.tail()
```

```python
test_df.head()
```

```python
X_train = train_df.drop(columns = 'SalePrice')
y_train = train_df['SalePrice']
```

```python
print("Shape of X_train dataframe: ", X_train.shape)
print("Shape of y_train dataframe: ", y_train.shape)
```

#  Numerical Missing value imputation

```python
num_vars = X_train.select_dtypes(include = ['int64', 'float64']).columns
print(num_vars)
```

```python
X_train[num_vars].isnull().sum()
```

```python
imputer_mean = SimpleImputer(strategy = 'mean')
# imputer_const = SimpleImputer(strategy = 'constant', fill_value = 99)


imputer_mean
```

```python
imputer_mean.fit(X_train[num_vars])
```

```python
imputer_mean.statistics_
```

```python
imputer_mean.transform(X_train[num_vars])
```

```python
X_train[num_vars] = imputer_mean.transform(X_train[num_vars])
# print(X_train[num_vars] )
```

```python
test_df[num_vars] = imputer_mean.transform(test_df[num_vars])
```

```python
X_train[num_vars].isnull().sum()
```

```python
test_df[num_vars].isnull().sum()
```

#  Categorical missing value imputation

```python
categ_vars = X_train.select_dtypes(include = ['O']).columns
categ_vars
```

```python
X_train[categ_vars].isnull().sum()
```

```python
imputer_mode = SimpleImputer(strategy = 'most_frequent')
```

```python
imputer_mode.fit(X_train[categ_vars])
```

```python
X_train[categ_vars] = imputer_mode.transform(X_train[categ_vars])
```

```python
test_df[categ_vars] = imputer_mode.transform(test_df[categ_vars])
```

```python
imputer_mode.statistics_
```

```python
X_train[categ_vars].isnull().sum()
```

```python
test_df[categ_vars].isnull().sum()
```
