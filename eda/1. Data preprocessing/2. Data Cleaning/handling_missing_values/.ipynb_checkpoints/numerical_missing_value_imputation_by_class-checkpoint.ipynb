```python
missing_values_20  =  ['Alley', 'MasVnrType', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature']
clear_datas = df.drop(columns = missing_values_20)
print(clear_datas.shape)
```

```python
numerical_datas = clear_datas.select_dtypes(include = ['int64', 'float64'])
numerical_datas.shape
```

```python
numerical_datas.isnull().sum()
```

```python
num_var_miss = ['LotFrontage','MasVnrArea','GarageYrBlt']
numerical_datas[num_var_miss][numerical_datas[num_var_miss].isnull().any(axis = 1)]
```

```python
df['LotConfig'].unique()
```

```python
df[df.loc[:, 'LotConfig'] == 'Inside']["LotFrontage"].replace(np.nan,df[df.loc[:, 'LotConfig'] == 'Inside']["LotFrontage"].mean() )
```

```python
### Mean

df_copy = df.copy()
num_vars_miss = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
categorical_vars = ['LotConfig', 'Exterior2nd', 'KitchenQual'] 

for cat_var, num_var_miss in zip(categorical_vars, num_vars_miss):
    if cat_var in df.columns and num_var_miss in df.columns:
        print(f"Processing {cat_var} and {num_var_miss}")
        for var_class in df[cat_var].unique():
            print(f"Processing class: {var_class}")
            # Calculate mean value for the specific categorical class
            mean_value = df[df[cat_var] == var_class][num_var_miss].mean()
            # Update df_copy with mean values where the categorical variable matches var_class
            df_copy.loc[df[cat_var] == var_class, num_var_miss] = df[df[cat_var] == var_class][num_var_miss].replace(np.nan, mean_value)
    else:
        print(f"Column {cat_var} or {num_var_miss} does not exist.")
```

```python
df_copy[num_vars_miss].isnull().sum()
```

```python
plt.figure(figsize=(15, 10))
sns.set(style="whitegrid")

for i, var in enumerate(num_vars_miss):
    plt.subplot(2, 2, i + 1)
    sns.histplot(df[var], bins=20, kde=True, color='red', linewidth=2, label='Original', alpha=0.3, stat='density')
    sns.histplot(df_copy[var], bins=20, kde=True, color='green', linewidth=2, label='After Imputation', alpha=0.3, stat='density')
    plt.title(f'Distribution of {var}')
    plt.legend()

plt.tight_layout()
plt.show()
```

#  Using Median

```python
df_copy = df.copy()
num_vars_miss = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']

medians_df = {var: df[var].median() for var in num_vars_miss}
medians_df_copy = {var: df_copy[var].median() for var in num_vars_miss}
```

```python
plt.figure(figsize=(15, 10))
sns.set(style="whitegrid")

for i, var in enumerate(num_vars_miss):
    plt.subplot(2, 2, i + 1)
    sns.histplot(df[var], bins=20, kde=True, color='red', linewidth=2, label='Original', alpha=0.3, stat='density')
    sns.histplot(df_copy[var], bins=20, kde=True, color='green', linewidth=2, label='After Imputation', alpha=0.3, stat='density')
    
    plt.axvline(medians_df[var], color='red', linestyle='--', linewidth=2, label=f'Original Median: {medians_df[var]:.2f}')
    plt.axvline(medians_df_copy[var], color='green', linestyle='--', linewidth=2, label=f'Imputed Median: {medians_df_copy[var]:.2f}')
    
    plt.title(f'Distribution of {var}')
    plt.legend()

plt.tight_layout()
plt.show()
```
