#  Missing value imputation by mean, median

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# Load datasets 
df = pd.read_csv('train.csv')


pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

df.head(10)
```

```python
df.tail(10)
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
null_var = df.isnull().sum() / df.shape[0] * 100
print("Missing values in percentage:\n", null_var)
```

```python
drop_columns = null_var[null_var > 20].keys()
print("Columns dropped due to high missing values:\n", drop_columns)

```

```python
new_df = df.drop(columns=drop_columns)
print("Shape after dropping columns:", new_df.shape)
```

```python
new_df.select_dtypes(include = ['int64','float64'])
new_df.head()
```

```python
plt.figure(figsize=(16, 9))
sns.heatmap(new_df.isnull(), cbar=False, cmap='viridis')
plt.title("Heatmap of Missing Values")
plt.show()
```

```python
missing_var = new_df.columns[new_df.isnull().any()]
numeric_missing_var = [var for var in missing_var if new_df[var].dtype in ['int64', 'float64']]
print("Numeric columns with missing values:", numeric_missing_var)
```

```python
new_df.isnull().sum()
```

```python
numeric_missing_var = [var for var in missing_var if new_df[var].dtype in ['int64', 'float64']]
print(numeric_missing_var)
```

```python

# Data distribution flow
plt.figure(figsize=(10, 10))
sns.set(style="whitegrid")

for i, var in enumerate(numeric_missing_var):
    plt.subplot(2, 2, i + 1)
    sns.histplot(new_df[var], bins=20, kde=False, color='#DC143C')
    sns.kdeplot(new_df[var].dropna(), linewidth=5, color='#DC143C')

plt.tight_layout()
plt.show()
```

```python
#  using : mean,median, or mode
#  if normally distributed ===> use mean method 
```

```python
new_df_by_mean = new_df.copy()
new_df_by_mean[numeric_missing_var] = new_df_by_mean[numeric_missing_var].fillna(new_df_by_mean[numeric_missing_var].mean())
```

```python
print("Filled missing values with mean:")
print(new_df_by_mean[numeric_missing_var].isnull().sum())
```

```python
plt.figure(figsize=(10, 10))
sns.set()

for i, var in enumerate(numeric_missing_var):
    plt.subplot(2, 2, i + 1)
    sns.histplot(new_df[var], bins=20, kde=False, color='#DC143C')
    sns.kdeplot(new_df_by_mean[var], linewidth=5, color='#DC143C')

plt.tight_layout()
plt.show()
```
