```python
# Import tensorflow 2
import tensorflow as tf
```

```python
# Syntax: tf.compat.v1.placeholder(dtype, shape = none, name = none)

# placeholders are not executable immediately so we need to disable eager execution in TF 2 not in 1

tf.compat.v1.disable_eager_execution()
```

# Starting with Placeholder

```python
# Creating placeholder 
# tf.placeholder(dtype = tf.float32, shape = (400,400), name = "Demo_placeholder")
a= tf.compat.v1.placeholder(dtype = tf.float32, shape = (400,400), name = "Demo_placeholder")
a
```

```python
b= tf.compat.v1.placeholder(dtype = tf.float32, shape = (400,400), name = "Demo_placeholder_2")
b
```

# Perform mathematical operations 

```python
# Add two tf_matrices

sum = tf.add(a,b)
print(sum)
```

```python
# Creating numpy array 

import numpy as np 

ones_array = np.ones((400,400), np.float32)
ones_array
```

#  Execute tensorflow placeholder using session

```python
with tf.compat.v1.Session() as session:
    add_result = session.run(sum, feed_dict={a: ones_array, b: ones_array})
    print("Addition Result:\n", add_result)
```

```python
# Applying subtraction division and multiplication operations also 

mul = tf.multiply(a, b)
div = tf.divide(a, b)
sub = tf.subtract(a, b)
```

```python
with tf.compat.v1.Session() as session:
    mul_result = session.run(mul, feed_dict={a: ones_array, b: ones_array})
    print("Multiplication Result:\n", mul_result)

    div_result = session.run(div, feed_dict={a: ones_array, b: ones_array})
    print("Division Result:\n", div_result)

    sub_result = session.run(sub, feed_dict={a: ones_array, b: ones_array})
    print("Subtraction Result:\n", sub_result)
```

# Using Lists for operation 

```python
list1 = [2,4,6,8,10]
list2 = [1,3,5,7,9]
```

```python
# Converting lists to tensrs 

tf1 = tf.constant(list1, dtype = tf.float32)
tf2 = tf.constant(list2, dtype = tf.float32)

print(tf1)
print(tf2)
```

```python
# Applying operations (add,subtract, multiply, divide)

add_lists = tf.add(tf1,tf2)
sub_lists = tf.subtract(tf1,tf2)
mul_lists = tf.multiply(tf1,tf2)
div_lists = tf.divide(tf1,tf2)
```

```python
with tf.compat.v1.Session() as session:
    add_list_result = session.run(add_lists)
    print("Addition Result (Lists):", add_list_result)

    mul_list_result = session.run(mul_lists)
    print("Multiplication Result (Lists):", mul_list_result)

    div_list_result = session.run(div_lists)
    print("Division Result (Lists):", div_list_result)

    sub_list_result = session.run(sub_lists)
    print("Subtraction Result (Lists):", sub_list_result)
```
