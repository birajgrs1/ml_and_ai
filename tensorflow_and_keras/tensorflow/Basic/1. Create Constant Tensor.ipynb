#  Create Constant numpy array 

```python
import numpy as np
```

```python
arr = tf.constant(np.array([1,2,3,4,5,6]))  # 1-D array 

arr
```

```python
# 2-D array 

arr_2D =  tf.constant([[1,2],[3,4]])

arr_2D
```

# Creating 1-D to 2-D

```python
arr_2D_1 =  tf.constant([1,2,3,4], shape = (2,2), dtype = "float32")

arr_2D_1
```

```python
#  Creating N-D array 

arrN = tf.constant([[1,2,3],[6,5,4],[10,100,1000]])
arrN
```

```python
arrN_1 = tf.constant([[1,2],[5,4],[100,1000]])
arrN_1
```
