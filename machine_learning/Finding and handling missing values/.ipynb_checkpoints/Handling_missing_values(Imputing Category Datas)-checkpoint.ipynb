#  Fill Values
### Syntax: df.fillna()

```python
df.fillna(method = "bfill", axis = 1)   #backward-filling
```

```python
df.fillna(method = "ffill", axis = 1)   #forward-filling  (columnwise ---axis =1, axis= 0 rowwise)
```

```python
#  mode-filling
df['Gender'].mode()[0]
df['Gender'].fillna(df['Gender'].mode()[0], inplace = True)
```

```python
df.head()
```

```python
# using loop to fill using mode
df.select_dtypes(include = "object").isnull().sum()
```

```python
for i in df.select_dtypes(include = "object").columns:
    df[i].fillna(df[i].mode()[0], inplace = True)
    # print(i)
```
