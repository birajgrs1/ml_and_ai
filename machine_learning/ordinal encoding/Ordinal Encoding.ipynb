```python
# importing libraries
import pandas as pd 
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.DataFrame({'size':['small', 'medium', 'large','extra large', 'medium', 'small', 'extra large',
'small', 'large', 'medium']})
df
```

```python
ord_data = [['small', 'medium', 'large', 'extra large']]
```

```python
# Applying ordinal encoder

oe = OrdinalEncoder(categories = ord_data)
oe.fit(df[['size']])
```

```python
oe.transform(df[['size']])
```

```python
df['size'] = oe.transform(df[['size']])
```
