```python
#  Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('../csv/loan.csv') 
df.head()
```

```python
df.isnull().sum()  #check null values
```

```python
# Check nature of datasets

sns.distplot(df['ApplicantIncome'])
plt.show()
```

```python
df.describe()
```

## Applying Scaling

```python
sc = StandardScaler()
```

```python
sc.fit(df[['ApplicantIncome']])
```

```python
df['ApplicantIncome_sc'] = sc.transform(df[['ApplicantIncome']])
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize = (10,7))
plt.subplot(1,2,1)
plt.title('Before')
sns.distplot(df[['ApplicantIncome_sc']])

plt.subplot(1,2,2)
plt.title('After')
sns.distplot(df[['ApplicantIncome']])
plt.show()
```
