```python
#  Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('../csv/loan.csv') 
df.head()
```

```python
df.isnull().sum()  #check null values
```

```python
# Check nature of datasets

sns.distplot(df['CoapplicantIncome'])
plt.show()
```

```python
df.describe()
```

### Applying Min-Max scaler(normalization)

```python
mmc = MinMaxScaler()
mmc.fit(df[['CoapplicantIncome']])
```

```python
df['CoapplicantIncome_mmc'] = mmc.transform(df[['CoapplicantIncome']])
```

```python
df.head()
```

```python
plt.figure(figsize = (12,6))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(df['CoapplicantIncome'])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(df['CoapplicantIncome_mmc'])
plt.show()
```
